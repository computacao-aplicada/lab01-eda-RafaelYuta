## Descobertas no dataset `winequality-red.csv`

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

wine = pd.read_csv('/content/winequality-red.csv',delimiter=";")
```

### 1. Distribuição da Qualidade dos Vinhos

```python
fig, ax = plt.subplots(figsize=(8, 5))
sns.countplot(x='quality', data=wine, palette='viridis', ax=ax)
ax.set_title('Distribuição da Qualidade dos Vinhos Tintos', fontsize=14, fontweight='bold')
ax.set_xlabel('Qualidade (0-10)')
ax.set_ylabel('Número de Vinhos')
plt.tight_layout()
plt.show()
```

A maioria dos vinhos tintos neste dataset recebe classificações de qualidade `5` e `6`. Isso sugere que o dataset é desbalanceado em relação à variável alvo 'quality'. Modelos de classificação podem ter um desempenho melhor nas classes majoritárias (5 e 6) e podem precisar de técnicas de reamostragem ou ponderação para melhorar a predição das classes minoritárias.

### 2. Relação entre Acidez Volátil e Qualidade

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.boxplot(x='quality', y='volatile acidity', data=wine, palette='coolwarm', ax=ax)
ax.set_title('Acidez Volátil por Qualidade do Vinho', fontsize=14, fontweight='bold')
ax.set_xlabel('Qualidade (0-10)')
ax.set_ylabel('Acidez Volátil')
plt.tight_layout()
plt.show()
```

Vinhos com menor acidez volátil tendem a ter classificações de qualidade mais altas, enquanto níveis elevados de acidez volátil estão associados a vinhos de menor qualidade. A acidez volátil em excesso é frequentemente um indicativo de deterioração e pode afetar negativamente o aroma e sabor do vinho. Esta evidência sugere que um controle rigoroso da acidez volátil é crucial para a produção de vinhos de alta qualidade.

### 3. Relação entre Ácido Cítrico e Qualidade

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.boxplot(x='quality', y='citric acid', data=wine, palette='viridis', ax=ax)
ax.set_title('Ácido Cítrico por Qualidade do Vinho', fontsize=14, fontweight='bold')
ax.set_xlabel('Qualidade (0-10)')
ax.set_ylabel('Ácido Cítrico')
plt.tight_layout()
plt.show()
```

Vinhos com maior teor de ácido cítrico parecem estar correlacionados com classificações de qualidade mais elevadas. Essa relação positiva sugere que o ácido cítrico é um componente desejável, indicando que sua presença em níveis adequados pode ser um fator importante para a percepção de uma maior qualidade.
